```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Data: Marketing budget (X) and Sales revenue (Y)
X = np.array([5, 10, 15, 20, 25]).reshape(-1, 1)  # Reshape for sklearn
Y = np.array([50, 65, 70, 80, 85])

# Create and train the Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predictions
Y_pred = model.predict(X)

# Calculate MSE and RMSE
MSE = mean_squared_error(Y, Y_pred)
RMSE = np.sqrt(MSE)

# Print regression line, MSE, and RMSE
print(f"Regression Line: Y = {model.coef_[0]}X + {model.intercept_}")
print(f"Mean Squared Error (MSE): {MSE}")
print(f"Root Mean Squared Error (RMSE): {RMSE}")

# Plot data points and regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color="blue", label="Actual Data (Points)")
plt.plot(X, Y_pred, color="red", label="Regression Line")

# Plot error lines
for i in range(len(X)):
    plt.vlines(x=X[i], ymin=Y_pred[i], ymax=Y[i], color="gray", linestyle="dashed", label="Error" if i == 0 else "")

# Annotate MSE and RMSE on the graph
plt.text(6, 60, f"MSE: {MSE:.2f}", fontsize=10, color="green")
plt.text(6, 55, f"RMSE: {RMSE:.2f}", fontsize=10, color="green")

# Labels and legend
plt.xlabel("Marketing Budget (in thousands)")
plt.ylabel("Sales Revenue (in thousands)")
plt.title("Linear Regression with Errors")
plt.legend()
plt.grid()
plt.show()
```
